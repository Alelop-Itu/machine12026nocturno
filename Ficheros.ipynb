# 1 Acceso a ficheros
## 1.1 Ficheros
* File es un tipo de objeto predefinido en Python (*built-in*).
* Permite acceder a ficheros desde programas Python.
* Los ficheros son de un tipo especial:
   * Son built-in, pero no son ni números, ni secuencias, ni mappings. Tampoco responden a operadores en expresiones.
* La función open permite crear objetos de tipo fichero.
  
Formato general para abrir un fichero:

afile = open(filename, mode)

* mode es opcional. Por defecto, los ficheros se abren en modo lectura.
* Los datos leidos de un fichero siempre se obtienen en formato string. Lo mismo ocurre con
escritura.
* Los ficheros se deben cerrar invocando close (liberación de recursos).

```python
# Lectura desde fichero usando método 'read'. Devuelve todo el contenido del fichero.
mi_fichero = open('res/multiple_lines.txt')
print(mi_fichero.read())
```

* Resolución de paths independiente de plataforma

```python
# Lectura linea a linea a través del bucle 'for'. Devuelve todo el contenido con espacio entre líneas

mi_fichero = open('res/multiple_lines.txt')
for linea in mi_fichero:
    print(f"{linea}")
mi_fichero.close()
```

```python
# leer de un archivo a una lista. Imprime en una sola oración , sin salto de línea

mi_fichero = open('res/multiple_lines.txt')
lineas = mi_fichero.readlines()
mi_fichero.close()

print(lineas)
```

* close autómatico con sentencia with. Esta es la forma habitual de leer de fichero en Python,
context manager

```python
# Lectura habitual de ficheros con cierre automático
with open(ruta) as mi_fichero:
    for linea in mi_fichero:
        print(linea, end = '')
```

```python
 #abrir varios ficheros en el mismo with
ruta1 = os.path.join("res", "one_line.txt") # Una sola oración, se imprime normal
ruta2 = os.path.join("res", "multiple_lines.txt") #3 oraciones, imprime con el símbolo de salto de línea

with open(ruta1) as fichero1, open(ruta2) as fichero2:
    print(fichero1.readlines()) 
    print(fichero2.readlines())
```

## 1.2 Modos de acceso
* Al crear un objeto de tipo File se puede espeficiar el modo de acceso (lectura/escritura).

| Modo de acceso | Desccripción |
|:---------|:-----|
| r | Solo Lectura |
| w | Solo Escritura (Borra si el archivo ya existe) |
| x | Solo Escritura (Falla si el archivo ya existe) |
| a | Crea Fichero (Si existe lo abre y se añade al final) |
| r+ | Lectura y Escritura |
| b | Se puede añadir a otros modos para acceso binario |
| t | Acceso para archivos de texto (default) |

## 1.3 Acceso para escritura

```python
def crear_lista(tamanyo):
    lista = []
    for i in range(tamanyo):
        lista.append(str(i) + '\n') #i se convierte en string y se agrega un salto de línea
    return lista

ruta = os.path.join("res", "a_dummy.txt") #Creación de un nuevo archivo en la ruta especificada
with open(ruta, 'wt') as fichero: #Abre el archivo en modo escritura
    fichero.write('Cabecera del ciclo del for en el ejemplo\n') #Escribe una línea inicial como encabezado
    lista = crear_lista(10) 
    fichero.writelines(lista) #Escribe todos los elementos de la lista en el archivo de una sola vez
    print("archivo creado")

```

**Buffering**
* Por defecto, el texto que transfieres desde tu programa a un fichero no se guarda en disco
inmediatamente. Se almacena en un buffer.
* Acciones como cerrar un fichero o invocar el método **flush** fuerzan que se transfiera el con
tenido del buffer a disco

```python
ruta = os.path.join("res", "FicheroParaEscritura.txt")
fichero_write = open(ruta, 'w')
fichero_write.write('foter') #Intenta escribir foter en el archivo pero se queda en buffer
fichero_read = open(ruta, "r") 
print(fichero_read.readlines())

fichero_write.flush() #Pasa del buffer al archivo creado
print(fichero_read.readlines()) 

fichero_write.close()
fichero_read.close()
```

## 1.4 Archivos CSV
* Python permite leer datos de ficheros CSV y también escribir ficheros en este formato.
* Popular formato en ciencia de datos

```python
# tabla_operaciones.csv contiene valores separados por comas
import csv #importa la librería
ruta = os.path.join("res", "tabla_operaciones.csv")

with open(ruta) as fichero:
    data_reader = csv.reader(fichero, delimiter=',') #Crea un objeto lector , interpreta el archivo mediante la coma
    for linea in data_reader:  # Itera sobre cada fila del archivo CSV.
        print(linea[0] + '---- ' + linea[1]) # Imprime el primer y segundo elemento de la fila actual, unidos por una cadena de guiones.
```

```python
ruta = os.path.join("res", "tabla_operaciones.csv") #Llama a la ruta de origen
f = open(ruta) #Abre el archivo de origen en modo lectura (sin usar 'with', por lo que requiere cierre manual).
data_reader = csv.reader(f, delimiter = ',') # Crea un lector para el archivo original usando la coma como separador.
ruta_o = os.path.join("res", "4_tabla_operaciones.csv")#Llama a la ruta a crear

with open(ruta_o, 'w') as f2: # Abre (o crea) el nuevo archivo en modo escritura.
    csv_writer = csv.writer(f2, delimiter = '|')
    for line in data_reader: #Recorre cada fila leída del archivo original.
            print(line)
            csv_writer.writerow(line) # Escribe la fila leída en el nuevo archivo, pero ahora utilizando el nuevo separador ('|').
f.close()  # Cierra manualmente el archivo de origen para liberar recursos.
```

## 2 Docstrings
* Python permite adjuntar documentación a los objetos e inspeccionarla a través de la línea de
comandos o durante la ejecución del programa.
* Los docstrings se almacenan en el atributo __doc__ de cada objeto.
* El valor de dicho atributo se puede mostrar por medio de la función *help*

```python
 help(open)
```

* No es necesario editar este atributo directamente.
* Para asociar un docstring a un objeto basta con escribir el texto (entre triples comillas) al
principio de los modulos, funciones o clases, antes del codigo ejecutable.

```python
def funcion_de_prueba():
    """Esta es la documentación de la función de prueba"""
    pass
help(funcion_de_prueba)
```

* Accediendo al atributo __doc__ sólo se obtiene el docstring

```python
print(funcion_de_prueba.__doc__)
```

Como se puede observar, *help* añade información adicional. Por ejemplo, para entidades más grandes, *help* muestra el docstring dividido en secciones.

```python
import sys
help(sys)
```

Los docstrings que ocupan más de una línea suelen tener estas partes:
* Resumen en la primera línea.
* Línea en blanco.
* Descripción detallada.
* Otra línea en blanco antes del código

```python
def funcion_de_prueba2():
    """Esta es la línea para el resumen.
    Este es el párrafo donde se puede escribir una descripción más
    detallada de la función. Observa que el resumen y la descripción
    detallada van separados por una línea en blanco. También hay
    otra línea en blanco antes de empezar el código.
    """
pass
help(funcion_de_prueba2)
```

Algunos comentarios adicionales:
* Comentarios con ‘#’ suelen asociarse a expresiones sencillas, instrucciones individuales o
pequeños bloques de código.
* Los docstrings son más apropiados para construcciones de más alto nivel: funciones, clases y
módulos.

1. Escribe una función que reciba una ruta de un fichero de texto y una cadena de caracteres a
buscar y determine si la cadena aparece en el fichero.

2. Escribe una función que reciba una lista, una ruta destino y un número n. La función debe
crear un fichero en la ruta especificada. El contenido del fichero serán los primeros n elementos
de la lista. La función debe controlar de manera apropiada los posibles valores de n que estén
fuera de rango.

```python
def buscar_en_fichero(ruta, cadena):
    with open(ruta, 'r') as fichero:
        for linea in fichero:
            if cadena in linea:
                return True
    return False

resultado = buscar_en_fichero("res/multiple_lines.txt", "hola")
print(resultado)
```

```python
def crear_fichero_con_lista(lista, ruta_destino, n):
    n_seguro = max(0,min(n, len(lista)))

    with open(ruta_destino, 'w') as f:
        for i in range(n_seguro):
            f.write(str(lista[i]) + '\n')

mi_lista = ["manzana", "pera", "uva", "naranja"]
crear_fichero_con_lista(mi_lista, "resultado.txt", 3)
print("Fichero creado correctamente")
```

3. Escribe una función que reciba una ruta de un fichero de texto devuelva un diccionario con la
frecuencia de aparición de cada palabra. Ejemplo: un fichero que contenga la frase ‘es mejor
que venga que que no venga’ devolverá el siguiente diccionario: {‘es’ : 1, ‘mejor’ : 1, ‘que’ :
3, ‘venga’ : 2, ‘no’ : 1}. Para dividir un string en palabras puedes hacer uso del método split.

```python
def frecuencia_palabras(ruta):
    frecuencia = {}
    with open(ruta, 'r') as f: #modo lectura
        contenido = f.read()
        palabras = contenido.split() # texto se divide en palabras con split
        
        for palabra in palabras:
            # Limpiamos posibles signos de puntuación básicos si fuera necesario
            palabra = palabra.strip('.,!¡¿?')
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
            
    return frecuencia

resultado = frecuencia_palabras("res/multiple_lines.txt")
print(resultado)
```
